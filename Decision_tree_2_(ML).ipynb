```python
# Import necessary libraries
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt

# Load the dataset
file_path = '/content/Enjoy sports (1).csv'  # Update this with the actual path to your CSV file
df = pd.read_csv(file_path)
```

```python
# Display the first few rows of the dataset (optional)
print("Original Dataset:")
print(df.head())
```

```python
# Apply Label Encoding to categorical columns
label_encoder = LabelEncoder()
```

```python
# Convert categorical features to numeric using LabelEncoder for all columns except 'Decision'
for column in df.columns[:-1]:  # We skip the 'Decision' column since it's the target
    df[column] = label_encoder.fit_transform(df[column])
```

```python
# Display the encoded dataset (optional)
print("\nEncoded Dataset:")
print(df.head())
```

```python
# Define features (X) and target (y)
X = df.drop(columns=['Decision'])  # Features are all columns except 'Decision'
y = df['Decision']  # Target is the 'Decision' column
```

```python
# Create and train the Decision Tree Classifier
clf = DecisionTreeClassifier(criterion='entropy')  # Using entropy criterion for splitting
clf.fit(X, y)
```

```python
# Plot and visualize the decision tree
plt.figure(figsize=(12, 8))
# Adjust figure size for better visibility
plot_tree(clf, feature_names=X.columns, class_names=['No', 'Yes'], filled=True, rounded=True, fontsize=10)
# Show the decision tree plot
plt.show()
```
