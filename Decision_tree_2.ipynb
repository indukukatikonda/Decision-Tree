```python
# Import necessary libraries
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt

# Load the dataset
file_path = '/content/Enjoy sports (1).csv'
df = pd.read_csv(file_path)


```

```python
# Convert categorical features to numeric using LabelEncoder
label_encoder = LabelEncoder()
for column in df.columns[:-1]:  # Exclude the target column 'Decision'
    df[column] = label_encoder.fit_transform(df[column])
```

```python
# Define features (X) and target (y)
X = df.drop(columns=['Decision'])  # Features are all columns except 'Decision'
y = df['Decision']  # Target is the 'Decision' column
```

```python
 #Create and train the Decision Tree Classifier
clf = DecisionTreeClassifier(criterion='entropy')  # Using entropy to match manual entropy calculation
clf.fit(X, y)
```

```python
# Plot and visualize the decision tree
plt.figure(figsize=(12, 8))
plot_tree(clf, feature_names=X.columns, class_names=['No', 'Yes'], filled=True, rounded=True)
plt.show()
```
